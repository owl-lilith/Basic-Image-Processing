```python
# implement libraries
import sys
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# global variable
MAX_KERNEL_LENGTH = 31
```

```python
# applying blur filters
image_path = 'test_input.png'
input_image = cv.imread(image_path)
output_image = np.copy(input_image)
```

```python
# normalized block filter (Homogeneous blur)
def normalized_block_filter(image):
    return cv.blur(image, (MAX_KERNEL_LENGTH, MAX_KERNEL_LENGTH))
```

```python
# gaussian filter
def gaussian_filter(image):
    return cv.GaussianBlur(image, (MAX_KERNEL_LENGTH, MAX_KERNEL_LENGTH), 0)
```

```python
# median filter
def median_filter(image):
    return cv.medianBlur(image, MAX_KERNEL_LENGTH)
```

```python
# bilateral filter
def bilateral_filter(image):
    return cv.bilateralFilter(image, MAX_KERNEL_LENGTH, MAX_KERNEL_LENGTH * 2, MAX_KERNEL_LENGTH / 2)
```

```python
# plot output images
def plot_images(images, images_title):
    plt.figure(figsize=(10, 6)) 
    for i in range(len(images)):
        plt.subplot(2, 2, i + 1) 
        img_rgb = cv.cvtColor(images[i], cv.COLOR_BGR2RGB)
        plt.imshow(img_rgb) 
        plt.title(images_title[i])  
        plt.axis('off')  # Hide the axes to make it cleaner

    plt.tight_layout()
    plt.show() 
```

```python
# main
images = [normalized_block_filter(input_image), gaussian_filter(input_image), median_filter(input_image), bilateral_filter(input_image)]
image_title = ['normalized block filter', 'gaussian filter', 'median filter', 'bilateral filter']
plot_images(images, image_title)
```
